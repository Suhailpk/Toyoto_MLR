# model building and training

```python
from sklearn.linear_model import LinearRegression
```

```python
X = toyota_corolla.drop(labels='Price',axis=1)
X
```

```python
y = toyota_corolla[['Price']]
y
```

```python
lin_model = LinearRegression()
```

```python
toyota_corolla.mean()
```

```python
X.std()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scalar = StandardScaler()
```

```python
scaled_X = std_scalar.fit_transform(X)
scaled_X
```

```python
scaled_X = pd.DataFrame(data=scaled_X,columns=["Age_08_04","KM","HP","cc","Doors","Gears","Quarterly_Tax","Weight"])
```

```python
scaled_X
```

```python
scaled_X.mean()
```

```python
scaled_X.std()
```

## split data train,test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.4, random_state=12)
```

```python
X_train
```

```python
lin_model.fit(X_train,y_train)
```

```python
y_pred_test = lin_model.predict(X_test)
y_pred_test
```

```python
error = y_test - y_pred_test
error
```
